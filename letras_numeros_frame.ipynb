```python
import pandas as pd

cliente = pd.read_csv('listagem_telefone.csv', sep=';', encoding ='latin1')
lista_cliente = []
lista_telefone = []
lista_telefone_codigo = []
```

```python
cliente
```

```python
for index, row in cliente.iterrows():
  verifica = (row['TELEFONE'])
  codigo =(row['CODIGO'])
  verdade = any(c.isalpha() for c in verifica)
  if (verdade == True):
    lista_telefone.append(verifica)
    lista_telefone_codigo.append(codigo)
  else:
    lista_cliente.append(verifica)


  
  

```

```python
lista_cliente
```

```python
lista_telefone
```

```python
lista_telefone_codigo
```

```python
lista_de_tuplas = list(zip(lista_telefone_codigo, lista_telefone))
```

```python
lista_de_tuplas
```

```python
saida_frame = pd.DataFrame(lista_de_tuplas, columns=['codigo', 'telefone'])
```

```python
saida_frame.to_csv('listagem_clientes_telefone_errado.csv', sep=';', index= None)
```
